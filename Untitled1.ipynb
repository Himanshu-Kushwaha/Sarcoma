```python
import numpy as np
import h5py
import scipy
from scipy.signal import wiener, medfilt
import pandas as pd
import matplotlib.pyplot as plt
from skimage.util.montage import montage2d
from skimage.color import label2rgb
from scipy import stats
import os
%matplotlib inline
```

```python
ct_data, pet_data, label_data = [], [], []
num_data = 1
with h5py.File(os.path.join('..','input','lab_petct_vox_5.00mm.h5'),'r') as p_data:
    for ct_img, pet_img, label in zip(p_data['ct_data'].values(),
                                      p_data['pet_data'].values(),
                                      p_data['label_data'].values()):
        ct_data += [np.array(ct_img)]
        pet_data += [np.array(pet_img)]
        label_data += [np.array(label)]
        num_data = num_data - 1
        if num_data == 0:
            break
#print(len(ct_data))
```

```python
vmin, vmax = stats.scoreatpercentile(ct_data[0], (0.5, 99.5))
```

```python
a = np.clip(ct_data[0], vmin, vmax)
a = (a - vmin) / (vmax - vmin)
plt.imshow(a[:,50,:][::-1,:],cmap='gray')
```

```python
from skimage import restoration
bilateral = restoration.denoise_bilateral(a)
```

```python
from skimage import exposure
hi_a = exposure.histogram(a)
hi_bilateral = exposure.histogram(bilateral)
plt.plot(hi_a[1], hi_a[0], label='ct_data')
plt.plot(hi_bilateral[1], hi_bilateral[0],
         label='bilateral')
plt.xlim(0, 0.5)
plt.legend()
plt.title('Histogram of voxel values')
```

```python
sample = bilateral > 0.1
plt.imshow(sample[:,50,:][::-1,:], cmap='gray')
```

```python
from scipy import ndimage
sample = ndimage.binary_fill_holes(sample)
plt.imshow(sample[:,50,:][::-1,:], cmap='gray')
```

```python
from skimage import morphology
open_object = morphology.opening(sample, morphology.ball(3))
close_object = morphology.closing(open_object, morphology.ball(3))
plt.imshow(close_object[:,50,:][::-1,:], cmap='gray')
```

```python
bbox = ndimage.find_objects(close_object)
```

```python
mask = close_object[bbox[0]]
```

```python
roi_dat = a[bbox[0]]
plt.imshow(roi_dat[:,50,:][::-1,:], cmap='gray')
plt.contour(mask[:,50,:][::-1,:])
print(roi_data.shape)
```

```python
nlm = restoration.denoise_nl_means(roi_dat[:, 
                                           20:80,
                                           20:80],
                    patch_size=5, patch_distance=7,
                    h=0.12, multichannel=False)
```

```python
plt.imshow(nlm[10], cmap='gray')
```

```python
tv = restoration.denoise_tv_chambolle(roi_dat[:, 20:80, 20:80],
                                      weight=0.2)
plt.imshow(tv[10], cmap='gray')
```

```python
hi_nlm = exposure.histogram(nlm)
hi_tv = exposure.histogram(tv)
```

```python
plt.plot(hi_nlm[1], hi_nlm[0])
plt.plot(hi_tv[1], hi_tv[0])
#plt.axvline(0.33, color='k', ls='--')
#plt.axvline(0.4, color='k', ls='--')
```

```python
markers = np.zeros(nlm.shape, dtype=np.uint8)
markers[nlm > 0.4] = 1
markers[nlm < 0.33] = 2
```

```python
plt.imshow(markers[10], cmap='gray')
plt.contour(markers[10], [0.5, 1.5])
```

```python
from skimage import segmentation
rw = segmentation.random_walker(nlm, markers, beta=1000., mode='cg_mg')
```

```python
plt.imshow(nlm[10], cmap='gray')
plt.contour(rw[10], [1.5])
```

```python
clean_segmentation = morphology.remove_small_objects(rw == 1, 200)
```

```python
from skimage import measure
labels = measure.label(clean_segmentation, 
                        background=0)
print(labels.max())
labels = (labels + 1).astype(np.uint8)
```

```python
from mayavi import mlab
```

```python
mlab.figure(bgcolor=(0, 0, 0))
s = mlab.pipeline.scalar_field(labels)
contour = mlab.pipeline.contour(s)
smooth = mlab.pipeline.user_defined(contour, filter='SmoothPolyDataFilter')
smooth.filter.number_of_iterations = 400
smooth.filter.relaxation_factor = 0.015
curv = mlab.pipeline.user_defined(smooth, filter='Curvatures')

surf = mlab.pipeline.surface(curv)
module_manager = curv.children[0]
module_manager.scalar_lut_manager.data_range = np.array([-0.6,  0.5])
module_manager.scalar_lut_manager.lut_mode = 'RdBu'
mlab.show()
```
